```python
def calculate_grade(marks):
    if marks >= 90 and marks <= 100:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    elif marks >= 0:
        return "F"
    else:
        return "Invalid Marks"
def student_report():
    students = {} 

    while True:
        name = input("Enter student name (or 'done' to finish): ").title()
        if name.lower() == "done":
            break  

        try:
            marks = float(input(f"Enter marks for {name} (0-100): "))
            grade = calculate_grade(marks)
            students[name] = grade
        except ValueError:
            print("‚ö†Ô∏è Please enter valid numeric marks.")
            continue

    print("\n===== üè´ Student Grade Report =====")
    for name, grade in students.items():
        print(f"{name}: {grade}")

    print("\nTotal students:", len(students))
    print("Grade")

student_report()
```
